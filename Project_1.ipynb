# Sorting

```python
# sorting the species alphabetically from largest to smallest
species_sorted_df = df.sort_values(by='species', ascending = False)
print(species_sorted_df)
```

```python
#sorting the length column from largest to smallest 
length_sorted_df = df.sort_values(by='length',ascending=False )
print(length_sorted_df)
```

```python
# sorting the species column descending and weight column ascending from largest to smallest
sw_sorted_df = df.sort_values(by=['species','weight'], ascending = [False, True])
print(sw_sorted_df)
```

```python
# sorting the w_l_ratio column from smallest to largest
wl_sorted_df = df.sort_values(by=['w_l_ratio'],ascending=True)
print(wl_sorted_df)
```
